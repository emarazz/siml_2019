# Data preparation

```
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

eps = np.finfo(float).eps

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Statistical inference/play_tennis.csv')
df=df.drop(['day'], axis=1)
df
```

# Functions 

```
def entrpy(df):

  entropy = 0
  column = df['play'].unique()
  for ii in column:
      norm_frequency = df['play'].value_counts()[ii]/len(df['play'])
      entropy =entropy -norm_frequency*np.log2(norm_frequency)
  return entropy


def entropy_feature(df,feature):
  '''
  entropy given feature
  '''
    
  yes_no = df['play'].unique()  
  sub_features = df[feature].unique()
  tot_entropy = 0


  for elmnt in sub_features:
    entropy = 0
    for yn in yes_no:
      freq = len(df[feature][df[feature]==elmnt][df['play'] ==yn])
      norm = len(df[feature][df[feature]==elmnt])
      norm_freq = freq/(norm+eps)
      entropy =entropy -norm_freq*np.log2(norm_freq+eps)

    prob = norm/14
    tot_entropy =tot_entropy -prob*entropy


  return abs(tot_entropy)

def best_IG(df):
    '''
    best feature.
    '''

    IG = []
    for key in df.keys()[:-1]:

        IG.append(entrpy(df)-entropy_feature(df,key))

    pos=np.argmax(IG)

    return df.keys()[pos]
```

# Decision Tree

```
# Find the root

root=best_IG(df)

print('Root is' ,root )


root_sub_features=df[root].unique()


# first layer

print('////////////Layer 1\\\\\\\\\\\\\\')
for value in root_sub_features:
        
  subtable = df[df[root] == value].reset_index(drop=True)
  clValue,counts = np.unique(subtable['play'],return_counts=True)                        
  
  if len(counts)==1:#Checking prob 0 or 1 condition
      print('Finalized Branch is' , value )                                                
      print('Decision of this branch is ',clValue[0])
      print('////////////Ended\\\\\\\\\\\\\\')
  else:
      print('Create branch called',value)
      print('////////////Continue\\\\\\\\\\\\\\')
      



# second layer
#  Cont' Sunny
print('                                     ')
print('////////////Layer 2- Sunny\\\\\\\\\\\\\\')
print('                                     ')
new_df = df[df[root] == 'Sunny'].reset_index(drop=True)


node1=best_IG(new_df)

print('Node of sunny branch is' ,node1 )


root_sub_features=df[node1].unique()


for value in root_sub_features:
        
  subtable = new_df[new_df[node1] == value].reset_index(drop=True)
  clValue,counts = np.unique(subtable['play'],return_counts=True)                        
  
  if len(counts)==1:#Checking prob 0 or 1 condition
      print('Finalized Branch is' , value )                                                
      print('Decision of this branch is ',clValue[0])
      print('////////////Ended\\\\\\\\\\\\\\')
  else:
      print('Create branch called',value)
      print('////////////Continue\\\\\\\\\\\\\\')




#  Cont' Rain
print('                                     ')
print('////////////Layer 2 - Rain\\\\\\\\\\\\\\')
print('                                     ')
new_df = df[df[root] == 'Rain'].reset_index(drop=True)


node2=best_IG(new_df)

print('Node of Rain branch is' ,node2 )


root_sub_features=df[node2].unique()


for value in root_sub_features:
        
  subtable = new_df[new_df[node2] == value].reset_index(drop=True)
  clValue,counts = np.unique(subtable['play'],return_counts=True)                        
  
  if len(counts)==1:#Checking prob 0 or 1 condition
      print('Finalized Branch is' , value )                                                
      print('Decision of this branch is ',clValue[0])
      print('////////////Ended\\\\\\\\\\\\\\')
  else:
      print('Create branch called',value)
      print('////////////Continue\\\\\\\\\\\\\\')
```
